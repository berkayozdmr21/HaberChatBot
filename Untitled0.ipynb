```python
# Gerekli kütüphaneler
import pandas as pd
import numpy as np
import torch
from transformers import BertTokenizerFast, BertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Google Drive'ı bağla (veri dosyanız orada ise)
from google.colab import drive
drive.mount('/content/drive')

```

```python
# Veri setini oku (dosya yolunu kendi Drive'ınızdaki konuma göre değiştirin)
df = pd.read_csv("/content/drive/My Drive/Nlp/Notebooks/dengelenmis_haber_veriseti.csv")

# Etiketleri bul ve sayısal hale getir
labels = df['Konu'].unique().tolist()
label_dict = {label: idx for idx, label in enumerate(labels)}
label_dict_inv = {v: k for k, v in label_dict.items()}

# 'Konu' sütununu sayısal hale çevir
df['label'] = df['Konu'].map(label_dict)

# Kontrol amaçlı
print(f"Etiketler ve indeksleri: {label_dict}")
print(f"Örnek veri:\n{df.head()}")
```

```python
# Metin ve etiketler
texts = df['Haber Başlığı'].tolist()
labels = df['label'].tolist()

# Train/Test ayır
train_texts, val_texts, train_labels, val_labels = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels)

# Tokenizer yükle
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')

# Tokenize işlemi
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(val_texts, truncation=True, padding=True, max_length=128)
```

```python
class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = NewsDataset(train_encodings, train_labels)
val_dataset = NewsDataset(val_encodings, val_labels)

```

```python
model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=len(label_dict))

```

```python
# Corrected TrainingArguments to use updated parameter names
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=50,
    weight_decay=0.01,
    eval_strategy="epoch",  # Changed from evaluation_strategy
    logging_strategy="epoch", # Changed from logging_strategy
    report_to="none",  # W&B kapalı
    save_strategy="no"
)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
from transformers import Trainer # Import Trainer

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc}

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)
```

```python
def evaluate_model_with_trainer(trainer, dataset, dataset_name=""):
    predictions = trainer.predict(dataset)
    preds = np.argmax(predictions.predictions, axis=1)
    labels = predictions.label_ids

    accuracy = accuracy_score(labels, preds)
    report = classification_report(
        labels,
        preds,
        target_names=[label_dict_inv[i] for i in range(len(label_dict))],
        zero_division=0
    )
    cm = confusion_matrix(labels, preds)

    print(f"\n========== {dataset_name} Seti Başarım Metrikleri ==========")
    print(f"Doğruluk (Accuracy): {accuracy:.4f}")
    print("\nSınıflandırma Raporu:\n")
    print(report)
    print("Confusion Matrix:")
    print(cm)

    # Dosyaya yaz
    save_path = f"/content/drive/My Drive/Nlp/Notebooks/{dataset_name.lower()}_metrics.txt"
    with open(save_path, "w", encoding="utf-8") as f:
        f.write(f"{dataset_name} Seti Başarım Metrikleri\n")
        f.write(f"Accuracy: {accuracy:.4f}\n\n")
        f.write("Sınıflandırma Raporu:\n")
        f.write(report + "\n\n")
        f.write("Confusion Matrix:\n")
        f.write(str(cm))
    print(f">>> Metrikler kaydedildi: {save_path}")

# Eğitim ve doğrulama setlerini değerlendir
evaluate_model_with_trainer(trainer, train_dataset, "Eğitim")
evaluate_model_with_trainer(trainer, val_dataset, "Doğrulama")
```

```python
trainer.train()
```

```python
predictions = trainer.predict(val_dataset)
y_true = val_labels
y_pred = np.argmax(predictions.predictions, axis=1)

print("========== DOĞRULAMA Seti Başarım Metrikleri ==========")
print("Doğruluk (Accuracy): {:.4f}".format(accuracy_score(y_true, y_pred)))
print("\nSınıflandırma Raporu:\n")
print(classification_report(y_true, y_pred, target_names=label_dict.keys()))
print("Confusion Matrix:")
print(confusion_matrix(y_true, y_pred))

# Sonuçları kayıt et
with open("/content/drive/My Drive/Nlp/Notebooks/doğrulama_metrics.txt", "w", encoding="utf-8") as f:
    f.write("Accuracy: {:.4f}\n\n".format(accuracy_score(y_true, y_pred)))
    f.write("Classification Report:\n")
    f.write(classification_report(y_true, y_pred, target_names=label_dict.keys()))
    f.write("Confusion Matrix:\n")
    f.write(str(confusion_matrix(y_true, y_pred)))
print(">>> Metrikler kaydedildi.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
import numpy as np
import pandas as pd

# Veriyi yükle
df = pd.read_csv("dengelenmis_haber_veriseti.csv")

# Etiketleri sayıya çevir
le = LabelEncoder()
y = le.fit_transform(df['Konu'])  # y sayısal (int) hale geldi
label_dict = dict(zip(le.classes_, le.transform(le.classes_)))  # Geri dönmek istersen diye

# TF-IDF ile haber başlıklarını sayısal vektörlere çevir
tfidf = TfidfVectorizer(max_features=1000)  # 1000 kelimeyle sınırlayabilirsin
X = tfidf.fit_transform(df['Haber Başlığı']).toarray()

# Eğitim ve test kümelerine ayır
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modeli kur
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(len(np.unique(y)), activation='softmax'))

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Eğit
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=16, verbose=1)

# Tahmin yap
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

# Sonuçlar
print(classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

accuracy = model.evaluate(X_test, y_test, verbose=0)[1]
print("\nAccuracy: %.2f%%" % (accuracy * 100))
```

```python
# Gerekli kütüphaneler
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

# 1. Veriyi yükle (CSV dosyanızla aynı dizinde olmalı)
df = pd.read_csv("dengelenmis_haber_veriseti.csv")  # Gerekirse yolu değiştirin

# 2. Metinleri temizle (küçük harfe çevir, noktalama ve sayıları kaldır)
def temizle(metin):
    metin = metin.lower()
    metin = re.sub(r'[^\w\s]', '', metin)  # Noktalama işaretlerini kaldır
    metin = re.sub(r'\d+', '', metin)      # Sayıları kaldır
    return metin

df['Haber Başlığı'] = df['Haber Başlığı'].apply(temizle)

# 3. Etiketleri sayısal değerlere dönüştür
le = LabelEncoder()
y = le.fit_transform(df['Konu'])  # 'Konu' sütunundaki sınıfları encode et
label_dict = dict(zip(le.classes_, le.transform(le.classes_)))  # İsteğe bağlı olarak sınıf isimlerini saklayalım

# 4. Türkçe stopword listesi
turkce_stopwords = [
    've', 'bir', 'bu', 'da', 'de', 'ile', 'için', 'ama', 'fakat', 'çok', 'daha', 'mi',
    'mu', 'mü', 'ben', 'sen', 'o', 'biz', 'siz', 'onlar', 'şu', 'ne', 'niçin', 'neden',
    'gibi', 'her', 'şey', 'ya', 'hem', 'hiç', 'en', 'ki', 'ise', 'ya da', 'ancak', 'bile',
    'diye', 'eğer', 'ise', 'mı', 'mu', 'mü', 'ile', 'olarak', 'bunun', 'bunlar', 'şimdi'
]

# 5. TF-IDF ile özellik çıkarımı (Unigram + Bigram, 3000 kelimeyle sınırlı)
tfidf = TfidfVectorizer(
    max_features=3000,
    ngram_range=(1, 2),
    stop_words=turkce_stopwords
)
X = tfidf.fit_transform(df['Haber Başlığı']).toarray()

# 6. Eğitim ve test setine ayır
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 7. Modeli tanımla (daha güçlü mimari)
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(np.unique(y)), activation='softmax'))

# 8. Derleme (compile)
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 9. Eğitimi başlat
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=25,
    batch_size=32,
    verbose=1
)

# 10. Tahminler ve sonuçlar
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

# 11. Performans çıktıları
print("\nSınıflandırma Raporu:\n")
print(classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0))

print("\nKarışıklık Matrisi:\n", confusion_matrix(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print("\nDoğruluk (Accuracy): %.2f%%" % (accuracy * 100))

# 12. Sonucu metin dosyasına yaz
from google.colab import drive
drive.mount('/content/drive')  # Drive'ı bağla (ilk çalıştırmada izin ister)

# Metni oluştur
sonuc_metni = "\nSınıflandırma Raporu:\n"
sonuc_metni += classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0)
sonuc_metni += "\n\nKarışıklık Matrisi:\n"
sonuc_metni += str(confusion_matrix(y_test, y_pred))
sonuc_metni += "\n\nDoğruluk (Accuracy): %.2f%%" % (accuracy * 100)

# Dosya yolunu tanımla (Drive içindeki 'Nlp/Notebooks' klasörüne)
dosya_yolu = "/content/drive/MyDrive/Nlp/Notebooks/sonuc.txt"

# Yazdır
with open(dosya_yolu, "w", encoding="utf-8") as f:
    f.write(sonuc_metni)

print(f"\nSonuçlar başarıyla '{dosya_yolu}' konumuna yazıldı.")
```

```python
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

# 1. Veriyi yükle
df = pd.read_csv("dengelenmis_haber_veriseti.csv")

# 2. Metin temizleme
def temizle(metin):
    metin = metin.lower()
    metin = re.sub(r'[^\w\s]', '', metin)
    metin = re.sub(r'\d+', '', metin)
    return metin

df['Haber Başlığı'] = df['Haber Başlığı'].apply(temizle)

# 3. Etiketleme
le = LabelEncoder()
y = le.fit_transform(df['Konu'])

# 4. Stopword listesi
turkce_stopwords = [
    've', 'bir', 'bu', 'da', 'de', 'ile', 'için', 'ama', 'fakat', 'çok', 'daha', 'mi',
    'mu', 'mü', 'ben', 'sen', 'o', 'biz', 'siz', 'onlar', 'şu', 'ne', 'niçin', 'neden',
    'gibi', 'her', 'şey', 'ya', 'hem', 'hiç', 'en', 'ki', 'ise', 'ya da', 'ancak', 'bile',
    'diye', 'eğer', 'ise', 'mı', 'mu', 'mü', 'ile', 'olarak', 'bunun', 'bunlar', 'şimdi'
]

# 5. TF-IDF
tfidf = TfidfVectorizer(
    max_features=3000,
    ngram_range=(1, 2),
    stop_words=turkce_stopwords
)
X = tfidf.fit_transform(df['Haber Başlığı']).toarray()

# 6. Veri bölme
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 7. Model oluştur
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(np.unique(y)), activation='softmax'))

# --- AŞAMA 1: Ana eğitim ---
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train,
          validation_data=(X_test, y_test),
          epochs=25,
          batch_size=32,
          verbose=1)

# --- AŞAMA 2: 1. Fine-tuning (lr=0.0005) ---
model.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train,
          validation_data=(X_test, y_test),
          epochs=5,
          batch_size=32,
          verbose=1)

# --- AŞAMA 3: 2. Fine-tuning (lr=0.0001) ---
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train,
          validation_data=(X_test, y_test),
          epochs=5,
          batch_size=32,
          verbose=1)

# --- AŞAMA 4: 3. Fine-tuning (lr=0.00005) ---
model.compile(
    optimizer=Adam(learning_rate=0.00005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train,
          validation_data=(X_test, y_test),
          epochs=5,
          batch_size=32,
          verbose=1)

# --- TAHMİN ve DEĞERLENDİRME ---
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

rapor = classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0)
matris = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

# --- SONUÇ DOSYASI YAZMA ---
from google.colab import drive
drive.mount('/content/drive')

sonuc_metni = "\nFine-tuning Sonrası Sınıflandırma Raporu:\n"
sonuc_metni += rapor
sonuc_metni += "\n\nKarışıklık Matrisi:\n"
sonuc_metni += str(matris)
sonuc_metni += "\n\nDoğruluk (Accuracy): %.2f%%" % (accuracy * 100)

dosya_yolu = "/content/drive/MyDrive/Nlp/Notebooks/sonuc2.txt"

with open(dosya_yolu, "w", encoding="utf-8") as f:
    f.write(sonuc_metni)

print(f"\nSonuçlar başarıyla '{dosya_yolu}' konumuna yazıldı.")
```
